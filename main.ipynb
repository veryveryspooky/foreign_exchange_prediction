```python
import numpy as np
import pandas as pd
from fredapi import Fred
import matplotlib.pyplot as plt
```

## Import data

```python
fred = Fred('725b5a988dcd81a9828c82cf8906d571')
```

```python
usd_eur = pd.DataFrame(fred.get_series('DEXUSEU')).reset_index().rename(columns={'index':'Date', 0:'value'})
usd_jpy = pd.DataFrame(fred.get_series('DEXJPUS')).reset_index().rename(columns={'index':'Date', 0:'value'})
usd_jpy['value'] = 1/usd_jpy['value']
usd_cny = pd.DataFrame(fred.get_series('DEXCHUS')).reset_index().rename(columns={'index':'Date', 0:'value'})
usd_cny['value'] = 1/usd_cny['value']
usd_gbp = pd.DataFrame(fred.get_series('DEXUSUK')).reset_index().rename(columns={'index':'Date', 0:'value'})
```

## Data explore

### USD/EUR

```python
plt.plot(usd_eur['Date'], usd_eur['value'])
plt.xlabel('Date')
plt.ylabel('USD/EUR')
plt.title('USD vs EUR')
```

```python
usd_eur.describe()
```

### USD/JPY

```python
plt.plot(usd_jpy['Date'], usd_jpy['value'])
plt.xlabel('Date')
plt.ylabel('USD/JPY')
plt.title('USD vs JPY')
```

### USD/CNY

```python
plt.plot(usd_cny['Date'], usd_cny['value'])
plt.xlabel('Date')
plt.ylabel('USD/CNY')
plt.title('USD vs CNY')
```

```python
plt.plot(usd_gbp['Date'], usd_gbp['value'])
plt.xlabel('Date')
plt.ylabel('USD/GBP')
plt.title('USD vs GBP')
```
