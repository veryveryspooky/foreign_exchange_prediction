```python
import numpy as np
import pandas as pd
from fredapi import Fred
import matplotlib.pyplot as plt
```

## Import data

```python
fred = Fred('725b5a988dcd81a9828c82cf8906d571')
```

```python
usd_eur = pd.DataFrame(fred.get_series('DEXUSEU')).reset_index().rename(columns={'index':'Date', 0:'value'})
jpy_usd = pd.DataFrame(fred.get_series('DEXJPUS')).reset_index().rename(columns={'index':'Date', 0:'value'})
cny_usd = pd.DataFrame(fred.get_series('DEXCHUS')).reset_index().rename(columns={'index':'Date', 0:'value'})
usd_gbp = pd.DataFrame(fred.get_series('DEXUSUK')).reset_index().rename(columns={'index':'Date', 0:'value'})
```

## Data explore

```python
plt.plot(usd_eur['Date'], usd_eur['value'])
plt.xlabel('Date')
plt.ylabel('USD/EUR')
```
